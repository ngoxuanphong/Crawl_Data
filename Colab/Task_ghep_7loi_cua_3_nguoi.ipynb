```python
df_merge_vang = pd.read_csv('/content/drive/MyDrive/Data Lake/Ingestion/Day 0/CafeF/Other/RequestByTrang/Compare/CompareScanP1.csv')
df_merge_vang = df_merge_vang.replace({-np.inf: 0, 0:np.nan})
df_merge_vang = df_merge_vang.drop(columns = ['Unnamed: 0'])
```

```python
df_merge_vang['cafef1'] = df_merge_vang['cafef'].apply(lambda val: val if val == np.nan else (round((round((val - 0.00001)/1000) - 0.00001)/1000) if val < 0 else round((round((val + 0.00001)/1000) + 0.00001)/1000)))
```

```python
df_merge_vang['cafef1'] = df_merge_vang['cafef'].apply(lambda val: round((round((val - 0.00001)/1000) - 0.00001)/1000) if val < 0 else round((round((val + 0.00001)/1000) + 0.00001)/1000))
df_merge_vang['VS1'] = df_merge_vang['VS'].apply(lambda val: round((val - 0.00001)/1000) if val < 0 else round((val + 0.00001)/1000))
df_merge_vang['Scan1'] = df_merge_vang['Scan'].apply(lambda val: 0 if val in [float("-inf"),float("inf")] else val)
df_merge_vang['Scan1'] = df_merge_vang['Scan1'].apply(lambda val: round((round((val - 0.00001)/1000) - 0.00001)/1000) if val < 0 else round((round((val + 0.00001)/1000) + 0.00001)/1000))
```

```python
def sosanhab(a,b):
    if a == b:
        return 1
    elif a*b != 0:
        return 0
    elif a == 0 and b == 0:
        return 1
    else: return 'N'
```

```python
df_merge_vang['Scan_cafef_1'] = df_merge_vang[['cafef1', 'Scan1']].apply(lambda x : sosanhab(x.cafef1, x.Scan1), axis = 1)
df_merge_vang['Scan_VS_1'] = df_merge_vang[['VS1', 'Scan1']].apply(lambda x : sosanhab(x.VS1, x.Scan1), axis = 1)
df_merge_vang['Scan_68_1'] = df_merge_vang[['68', 'Scan1']].apply(lambda x : sosanhab(x['68'], x.Scan1), axis = 1)
df_merge_vang.to_csv()
```
